```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
import lmfit
from lmfit.models import VoigtModel

from dataclasses import dataclass
from typing import List, Tuple, Protocol
import os.path
```

```python
def spectrum(file, prom=40, plot=False) -> np.ndarray:
    """For peak finding in clean spectra, e.g. strong calibration"""
    with open(file, 'r') as spe:
        bin_vals = spe.readlines()[12:-14]
        if len(bin_vals) != 2048:
            raise ValueError('Expected 2048 bin file; verify file formatting works.')
        bin_vals = np.array(bin_vals, dtype='int')
    x = np.linspace(0, 2047, 2048)
    p = scipy.signal.find_peaks(bin_vals, prominence=max(bin_vals)/prom)
    if plot:
        plt.plot(x, bin_vals, marker='o', linestyle='')
        plt.plot(p[0], bin_vals[p[0]], marker='x', linestyle='')
        plt.show()    
    return np.array(p[0])
```

```python
# let's plot some spectra
directory = '2022-10-31'
for path in os.listdir(directory):
    filepath = os.path.join(directory, path)
    print(filepath)
    spectrum(filepath, plot=True)
```

```python
@dataclass
class CalibrationFile:
    """Spectrum calibration file and its expected energies of peaks"""
    filepath: str
    energies: list
    rebinned: List[int] = None

@dataclass
class DataFile:
    """Spectrum file and scattering angle"""
    filepath: str
    angle: float
    rebinned: List[int] = None

class Detector:
    def __init__(self, name, cal_files: List[CalibrationFile], dat_files: List[DataFile], rebinning=8):
        self.name = name
        self.calfiles = cal_files
        self.datafiles = dat_files
        self.rebinning = rebinning
        self.angles = [f.angle for f in dat_files] # to store angle values
        self.peaks = [None for f in dat_files] # to store relevant peak values of angles
        self.cal = None
        
    def rebin(self, file) -> List[int]:
        if file.rebinned is not None:
            return file.rebinned
        # grab data
        with open(file.filepath, 'r') as spe:
            bin_vals = spe.readlines()[12:-14]
            if len(bin_vals) != 2048:
                raise ValueError('Expected 2048 bin file; verify file formatting works.')
            bin_vals = np.array(bin_vals, dtype='int')
        # rebin data
        newbins = []
        ct = 0
        for i, val in enumerate(bin_vals):
            ct += val
            if (i + 1) % self.rebinning == 0:
                newbins.append(ct)
                ct = 0
        return newbins
        
    def naive_calibrate(self, criterion=40) -> float:
        """Use the highest peak, correspond to the highest energy"""
        calvals = []
        for cf in self.calfiles:
            cr2 = criterion
            peak_idx = spectrum(cf.filepath, prom=criterion, plot=False)
            energy = max(cf.energies)
            
            while len(peak_idx) > len(cf.energies) * 1.5:
                # lower criterion until it works
                cr2 -= 1
                peak_idx = spectrum(cf.filepath, prom=cr2, plot=False)
                if cr2 < 1:
                    raise ValueError("Could not find peaks with reasonable prominence")
                    
            calvals.append(float(energy / peak_idx[-1]))
        self.cal = sum(calvals) / len(calvals)
        return self.cal

    def linear_calibrate(self, criterion=40, plot=False) -> float:
        """Try to fit the peak positions to the energies via a modified least-squares.
        If only one energy expected, defaults to naive_calibrate"""
        calvals = []
        for cf in self.calfiles:
            cr2 = criterion
            peak_idx = spectrum(cf.filepath, prom=criterion, plot=False)
            expect_energies = np.array(sorted(cf.energies))
            
            if len(expect_energies) < 2:
                print("Defaulting to naive fit for ", cf.filepath)
                calvals.append(expect_energies[-1] / peak_idx[-1])
                continue
            
            while len(peak_idx) > len(cf.energies) * 1.5:
                # lower criterion until it works
                cr2 -= 1
                peak_idx = spectrum(cf.filepath, prom=cr2, plot=False)
                if cr2 < 2:
                    raise ValueError("Could not find peaks with reasonable prominence")

            # match peaks to expected, with a scaling factor
            def scale_sq(factor):
                pred_energies = peak_idx * factor
                # for each peak, find closest expected (duplicate possible)
                x = [] # predictions from scaling observed peaks
                y = [] # prediction from known spectrum
                for e_pred in pred_energies:
                    x.append(e_pred)
                    diffs = (expect_energies - e_pred)**2
                    y.append(expect_energies[np.argmin(diffs)])
                return np.sum((np.array(x)-np.array(y))**2)

            # find optimum scale factor to minimize scale_sq
            res = scipy.optimize.minimize(scale_sq, 0.5, method='BFGS')
            calvals.append(res.x[0])
            
            if plot:
                pred_energies = peak_idx * res.x[0]
                x = []
                y = []
                for e_pred in pred_energies:
                    x.append(e_pred)
                    diffs = (expect_energies - e_pred)**2
                    y.append(expect_energies[np.argmin(diffs)])
                plt.plot(x, y)
                print("Scaled Bins", x)
                print("Expected Energies", y)
                plt.plot((0,1000), (0, 1000))
                plt.show()
                
            
        self.cal = sum(calvals) / len(calvals)
        return self.cal
    
    def plot_spectra(self):
        """Plot the spectrum, but add together self.rebinning adjacent bins."""
        assert self.cal is not None
        for f in self.datafiles:
            newbins = self.rebin(f)
            
            x = np.array([i for i in range(len(newbins))])
            plt.title(f"{f.angle} degree, {self.name}")
            plt.xlabel("Energy (keV)")
            plt.ylabel("Count")
            plt.plot(x * self.rebinning * self.cal, newbins, marker='o', linestyle='')
            plt.show()
            
    def fit_two_peaks(self, angle, guess1=100, guess2=400):
        assert self.cal is not None
        for f in self.datafiles:
            if f.angle != angle:
                continue
            # get data
            y = np.array(self.rebin(f))
            x = np.array([i for i in range(len(y))]) * self.rebinning * self.cal
            # trim zero bins
            mask = y != 0
            y = y[mask]
            x = x[mask]
            wt = 1 / np.sqrt(y)
            # fit
            peakA = VoigtModel(prefix='A_')
            pars = peakA.make_params(A_amplitude=10000, A_sigma=30, A_center=guess1)
            peakB = VoigtModel(prefix='B_')
            pars.update(peakB.make_params(B_amplitude=20000, B_sigma=30, B_center=guess2))
            model = peakA + peakB
            result = model.fit(y, params=pars, x=x, weights=wt, scale_covar=False)
            print(result.fit_report())
            result.plot()
        
```
